```python
import mysql.connector
def runquery(query):
    mydb = mysql.connector.connect(
        host="localhost",
        user="root",
        password="password",
        database="information_schema"
    )

    mycursor = mydb.cursor()
    mycursor.execute(query)

    for x in mycursor:
      print(x)
    mydb.close()
    mycursor.close()
```

```python
# PADA TABEL PAYMENTS DAN CUSTOMERS YG MENJD FIELD PEMERSATU ADALAH FIELD customerNumber
runquery2("SELECT customers.customerName, customers.creditLimit, payments.checkNumber, payments.amount FROM customers INNER JOIN payments ON customers.customerNumber = payments.customerNumber;")
```

INNER JOIN data yg ditampilkan adalah data yg di tabel 1 & tabel 2 ada nilainya (not null).
Perbedaan dari setiap join dapat dilihat pada gambar sql-joins & brkt adlh penjelsannya

1. INNER JOIN : data yang ditampilkan hanya data dimana pada tabel 1(kiri) ada datanya (not null) dan tabel 2(kanan) juga ada datanya (not null). waktu pemrosesan lebih lambat dari left dan right join

2. LEFT JOIN : data yang ditampilkan hanya data pada tabel yang disebelah kiri(tabel 1) saja yang ditampilkan sedangkan jika pada tabel kanan(2) bernilai NULL maka akan ditampilkan NULL. waktu pemrosesan left join lbh cpt dari inner join

3. RIGHT JOIN : data yang ditampilkan hanya data pada tabel yang disebelah kanan(tabel 2) saja yang ditampilkan sedangkan jika pada tabel kiri(1) bernilai NULL maka akan ditampilkan NULL. waktu pemrosesan right join lbh cpt dari inner join

4. FULL JOIN : data yang ditampilkan adalah semua data baik yang di tabel kanan(2) maupun kiri(1).

semua jenis join tersebut perlu field penghubung dimana field penghubung ini harus ada di tabel 1 dan tabel 2. Format query nya sama saja sebnrnya yaitu
* SELECT [APA YANG INGIN DITAMPILKAN] FROM TABEL1(KIRI) INNER || LEFT || RIGHT || FULL JOIN TABEL2(KANAN) ON [FIELD PENGHUBUNG DITABEL1(KIRI)] = [FIELD PENGHUBUNG DITABEL2(KANAN)

```python
# AGAR LEBIH MUDAH PENULISAN TABELNYA KITA GUNAKAN ALIAS
runquery2("SELECT CUS.customerName, CUS.creditLimit, PAY.checkNumber, PAY.amount FROM customers CUS LEFT JOIN payments PAY ON CUS.customerNumber = PAY.customerNumber;")
```

```python
runquery2("SELECT CUS.customerName, CUS.creditLimit, PAY.checkNumber, PAY.amount FROM customers CUS RIGHT JOIN payments PAY ON CUS.customerNumber = PAY.customerNumber;")
```

```python
# KARENA FIELD/COLUMN PENGHUBUNGNYA BERNAMA SAMA (customerNumber) MAKA KITA BISA GUNAKAN USING
runquery2("SELECT CUS.customerName, CUS.creditLimit, PAY.checkNumber, PAY.amount FROM customers CUS INNER JOIN payments PAY USING(customerNumber) ORDER BY CUS.creditLimit;")
```

```python
# disini mysql yg digunakan adalah versi 8 maka FULL JOIN tidak ada tapi diganti dengan UNION
runquery2("SELECT customerName, creditLimit FROM customers UNION SELECT checkNumber, amount FROM payments;")
```

```python
# UNION ALL
runquery2("SELECT customerName, creditLimit FROM customers  UNION ALL SELECT checkNumber, amount FROM payments ORDER BY creditLimit ASC;")
```
